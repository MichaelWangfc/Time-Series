```python
frontLight_left_2016 = pd.read_csv('2016Taibao-front-light.csv',encoding='gbk')
```

```python
#construct the dataframe with the index of date
frontlight = frontLight_left_2016[['CLAIM_AMOUNT','LOSS_AMOUNT']]
frontlight.set_index(frontLight_left_2016['ENDCASE_TIME'],inplace=True)
frontlight
```

```python
pd.to_datetime(frontlight.index.values)
```

```python
frontlight.set_index(pd.to_datetime(frontlight.index),inplace=True)
```

```python
frontlight.sort_index(inplace=True)
frontlight['Tick']=range(len(frontlight))
frontlight
```

```python
# find mean price in a given month
frontlight.shape
frontlight.groupby(pd.TimeGrouper(freq='M')).mean()
```

```python
month_groupby = frontlight.groupby(pd.TimeGrouper(freq='M'))
month_max = month_groupby['LOSS_AMOUNT'].max()
month_min = month_groupby['LOSS_AMOUNT'].min()
# month_max
month_difference = (month_groupby['LOSS_AMOUNT'].max() - month_groupby['LOSS_AMOUNT'].min())
month_mean = month_groupby['LOSS_AMOUNT'].mean()
month_index= month_max.index
data = np.array([month_min,month_max,month_difference,month_mean]).transpose()


month_analysis = pd.DataFrame(data)

month_analysis.columns= ['Monthly LOSS_AMOUNT min','Monthly LOSS_AMOUNT max','Monthly LOSS_AMOUNT Variation','Monthly LOSS_AMOUNT mean']
month_analysis.set_index(month_index,inplace=True)
month_analysis.head()
```

```python
month_analysis.reset_index(inplace=True)
```

```python
# plot the time serise in 2016
# month_analysis.columns
month_analysis.plot('ENDCASE_TIME','Monthly LOSS_AMOUNT mean')
```
